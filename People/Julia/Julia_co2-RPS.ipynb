```python
import pandas as pd
file='./data/ICO2sensordata_asc.txt'
```

```python
df = pd.read_csv(file,skiprows = [0,1],parse_dates = [[0,1]], index_col = 0, sep = r"\s*", 
    names=['date','time','CO2','Temperature','Humidity','Relative Humidity','Pressure'],header = None)
```
